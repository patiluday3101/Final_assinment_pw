```python
3)
class LibraryBook:
    def __init__(self, book_name, author, availability_status=True):
        # Encapsulating the attributes by making them private
        self.__book_name = book_name
        self.__author = author
        self.__availability_status = availability_status

    # Getter methods to access private attributes
    def get_book_name(self):
        return self.__book_name

    def get_author(self):
        return self.__author

    def is_available(self):
        return self.__availability_status

    # Setter methods to modify private attributes
    def set_book_name(self, book_name):
        self.__book_name = book_name

    def set_author(self, author):
        self.__author = author

    def set_availability_status(self, status):
        self.__availability_status = status

    # Method to borrow the book
    def borrow_book(self):
        if self.__availability_status:
            self.__availability_status = False
            print(f"You have borrowed '{self.__book_name}' by {self.__author}.")
        else:
            print(f"Sorry, '{self.__book_name}' by {self.__author} is not available right now.")

    # Method to return the book
    def return_book(self):
        if not self.__availability_status:
            self.__availability_status = True
            print(f"Thank you for returning '{self.__book_name}' by {self.__author}.")
        else:
            print(f"'{self.__book_name}' by {self.__author} was not borrowed.")


# Testing the LibraryBook class
book1 = LibraryBook("The Great Gatsby", "F. Scott Fitzgerald")
book2 = LibraryBook("1984", "George Orwell", availability_status=False)

# Displaying book details and borrowing books
print(f"Book: {book1.get_book_name()} by {book1.get_author()} - Available: {book1.is_available()}")
book1.borrow_book()

print(f"Book: {book2.get_book_name()} by {book2.get_author()} - Available: {book2.is_available()}")
book2.borrow_book()

# Returning books
book1.return_book()
book2.return_book()
```

```python
4)
# Base class for BankAccount
class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    # Method for deposit
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    # Method for withdrawal
    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient balance.")

    # Method to check balance
    def get_balance(self):
        print(f"Balance: {self.balance}")
        return self.balance

# Subclass for Savings Account
class SavingsAccount(BankAccount):
    def __init__(self, account_holder, balance=0, interest_rate=0.02):
        super().__init__(account_holder, balance)
        self.interest_rate = interest_rate

    # Method to apply interest
    def apply_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Applied interest of {interest}. New balance: {self.balance}")

# Subclass for Checking Account
class CheckingAccount(BankAccount):
    def __init__(self, account_holder, balance=0, overdraft_limit=500):
        super().__init__(account_holder, balance)
        self.overdraft_limit = overdraft_limit

    # Override withdraw method to allow overdraft
    def withdraw(self, amount):
        if amount > 0 and (self.balance + self.overdraft_limit) >= amount:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Insufficient funds, even with overdraft protection.")

# Test the banking system

        # Create instances of different account types
savings_account = SavingsAccount("Alice", 1000)
checking_account = CheckingAccount("Bob", 500)

        # Perform transactions on Savings Account
savings_account.deposit(200)
savings_account.withdraw(100)
savings_account.apply_interest()
savings_account.get_balance()

        # Perform transactions on Checking Account
checking_account.deposit(300)
checking_account.withdraw(600)
checking_account.withdraw(1200)
checking_account.get_balance()
```

```python
6)
# Base class for MenuItem
class MenuItem:
    def __init__(self, item_id, name, description, price, category):
        self.__item_id = item_id  # Encapsulation: item_id is hidden
        self.name = name
        self.description = description
        self.price = price
        self.category = category

    # Getter for item_id
    def get_item_id(self):
        return self.__item_id

    # Method to update menu item information
    def update_item(self, name=None, description=None, price=None, category=None):
        if name:
            self.name = name
        if description:
            self.description = description
        if price:
            self.price = price
        if category:
            self.category = category
        print(f"Menu item '{self.name}' updated.")

    # Method to display menu item details
    def display_item(self):
        print(f"ID: {self.__item_id}, Name: {self.name}, Description: {self.description}, Price: {self.price}, Category: {self.category}")

    # Method to remove menu item
    def remove_item(self):
        print(f"Menu item '{self.name}' removed.")


# Subclass for FoodItem
class FoodItem(MenuItem):
    def __init__(self, item_id, name, description, price, category, food_type):
        super().__init__(item_id, name, description, price, category)
        self.food_type = food_type  # Additional attribute specific to food items

    # Method to update food-specific information
    def update_food_info(self, food_type=None):
        if food_type:
            self.food_type = food_type
        print(f"Food item '{self.name}' updated with food type: {self.food_type}")

    # Method to display food item details
    def display_food_item(self):
        self.display_item()
        print(f"Food Type: {self.food_type}")


# Subclass for BeverageItem
class BeverageItem(MenuItem):
    def __init__(self, item_id, name, description, price, category, volume):
        super().__init__(item_id, name, description, price, category)
        self.volume = volume  # Additional attribute specific to beverages

    # Method to update beverage-specific information
    def update_beverage_info(self, volume=None):
        if volume:
            self.volume = volume
        print(f"Beverage item '{self.name}' updated with volume: {self.volume}ml")

    # Method to display beverage item details
    def display_beverage_item(self):
        self.display_item()
        print(f"Volume: {self.volume}ml")


# Testing the Restaurant Management System
if __name__ == "__main__":
    # Create food and beverage items
    pizza = FoodItem(1, "Pizza Margherita", "Classic pizza with cheese and tomato", 12.99, "Main Course", "Vegetarian")
    soda = BeverageItem(2, "Coke", "Carbonated soft drink", 2.99, "Beverage", 330)

    # Display food and beverage item details
    print("\nFood Item Details:")
    pizza.display_food_item()

    print("\nBeverage Item Details:")
    soda.display_beverage_item()

    # Update food item information
    pizza.update_item(price=13.99)
    pizza.update_food_info(food_type="Non-Vegetarian")

    # Display updated food item details
    print("\nUpdated Food Item Details:")
    pizza.display_food_item()

    # Remove a menu item
    pizza.remove_item()
```

```python
# Base class for Room
class Room:
    def __init__(self, room_number, room_type, rate, availability=True):
        self.__room_number = room_number  # Encapsulation: room_number is private
        self.room_type = room_type
        self.rate = rate
        self.__availability = availability

    # Getter for room_number
    def get_room_number(self):
        return self.__room_number

    # Getter for availability status
    def is_available(self):
        return self.__availability

    # Method to book a room
    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.get_room_number()} booked successfully.")
        else:
            print(f"Room {self.get_room_number()} is not available.")

    # Method to check in a guest
    def check_in(self, guest_name):
        if not self.__availability:
            print(f"Guest {guest_name} checked into Room {self.get_room_number()}.")
        else:
            print(f"Room {self.get_room_number()} is available. Please book the room first.")

    # Method to check out a guest
    def check_out(self, guest_name):
        if not self.__availability:
            self.__availability = True
            print(f"Guest {guest_name} checked out from Room {self.get_room_number()}.")
        else:
            print(f"Room {self.get_room_number()} is available. No check-out needed.")

# Subclass for SuiteRoom
class SuiteRoom(Room):
    def __init__(self, room_number, rate, num_bedrooms, has_jacuzzi):
        super().__init__(room_number, "Suite", rate)
        self.num_bedrooms = num_bedrooms  # Additional attribute for SuiteRoom
        self.has_jacuzzi = has_jacuzzi    # Additional attribute for SuiteRoom

    # Method to describe SuiteRoom
    def describe_suite(self):
        jacuzzi_info = "with jacuzzi" if self.has_jacuzzi else "without jacuzzi"
        print(f"Suite Room {self.get_room_number()} - {self.num_bedrooms} bedrooms, {jacuzzi_info}, Rate: {self.rate}")

# Subclass for StandardRoom
class StandardRoom(Room):
    def __init__(self, room_number, rate, has_air_conditioning):
        super().__init__(room_number, "Standard", rate)
        self.has_air_conditioning = has_air_conditioning  # Additional attribute for StandardRoom

    # Method to describe StandardRoom
    def describe_standard(self):
        ac_info = "with air conditioning" if self.has_air_conditioning else "without air conditioning"
        print(f"Standard Room {self.get_room_number()} - {ac_info}, Rate: {self.rate}")

# Testing the Hotel Management System
if __name__ == "__main__":
    # Create instances of SuiteRoom and StandardRoom
    suite = SuiteRoom(101, 250, 2, True)
    standard = StandardRoom(102, 100, True)

    # Describe rooms
    suite.describe_suite()
    standard.describe_standard()

    # Book rooms
    suite.book_room()
    standard.book_room()

    # Check in guests
    suite.check_in("John Doe")
    standard.check_in("Jane Smith")

    # Check out guests
    suite.check_out("John Doe")
    standard.check_out("Jane Smith")

    # Try to rebook and check-in again
    suite.book_room()  # Room already booked
    standard.check_in("Michael Johnson")  # Room not available for check-in
```

```python
# Base class for Member
class Member:
    def __init__(self, member_id, name, age, membership_type, membership_status=True):
        self.__member_id = member_id  # Private attribute: member_id is hidden using encapsulation
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__membership_status = membership_status  # Private attribute: membership_status is hidden

    # Getter for member_id (to access the private member_id)
    def get_member_id(self):
        return self.__member_id

    # Getter for membership status
    def get_membership_status(self):
        return self.__membership_status

    # Method to register a new member
    def register_member(self):
        if self.__membership_status:
            print(f"{self.name} is already a registered member.")
        else:
            self.__membership_status = True
            print(f"{self.name} has been successfully registered with {self.membership_type} membership.")

    # Method to renew the membership
    def renew_membership(self):
        if self.__membership_status:
            print(f"{self.name}'s membership has been renewed.")
        else:
            print(f"{self.name} needs to register before renewing the membership.")

    # Method to cancel the membership
    def cancel_membership(self):
        if self.__membership_status:
            self.__membership_status = False
            print(f"{self.name}'s membership has been cancelled.")
        else:
            print(f"{self.name} does not have an active membership to cancel.")


# Subclass for FamilyMember
class FamilyMember(Member):
    def __init__(self, member_id, name, age, membership_type, num_family_members):
        super().__init__(member_id, name, age, membership_type)
        self.num_family_members = num_family_members  # Additional attribute for family members

    # Method to update the number of family members
    def update_family_members(self, num_family_members):
        self.num_family_members = num_family_members
        print(f"{self.name} now has {self.num_family_members} family members added to the membership.")

    # Method to display family member details
    def display_family_member_details(self):
        print(f"Family Member: {self.name}, Age: {self.age}, Membership Type: {self.membership_type}, Family Members: {self.num_family_members}")


# Subclass for IndividualMember
class IndividualMember(Member):
    def __init__(self, member_id, name, age, membership_type, is_workout_personalized):
        super().__init__(member_id, name, age, membership_type)
        self.is_workout_personalized = is_workout_personalized  # Personalized workout option

    # Method to update personalized workout option
    def update_personalized_workout(self, is_workout_personalized):
        self.is_workout_personalized = is_workout_personalized
        print(f"{self.name}'s personalized workout preference has been updated to: {self.is_workout_personalized}")

    # Method to display individual member details
    def display_individual_member_details(self):
        print(f"Individual Member: {self.name}, Age: {self.age}, Membership Type: {self.membership_type}, Personalized Workout: {self.is_workout_personalized}")


# Testing the Fitness Club Management System
if __name__ == "__main__":
    # Create instances of FamilyMember and IndividualMember
    family_member = FamilyMember(1, "John Doe", 40, "Family", 3)
    individual_member = IndividualMember(2, "Jane Smith", 25, "Individual", True)

    # Display member details
    family_member.display_family_member_details()
    individual_member.display_individual_member_details()

    # Register members
    family_member.register_member()
    individual_member.register_member()

    # Update family member information
    family_member.update_family_members(4)

    # Update individual member workout preference
    individual_member.update_personalized_workout(False)

    # Renew memberships
    family_member.renew_membership()
    individual_member.renew_membership()

    # Cancel memberships
    family_member.cancel_membership()
    individual_member.cancel_membership()

    # Try re-canceling the membership
    family_member.cancel_membership()
```

```python
# Base class for Event
class Event:
    def __init__(self, event_id, name, date, time, location):
        self.__event_id = event_id  # Private attribute: event_id is hidden using encapsulation
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []  # Private attribute: list of attendees

    # Getter for event_id (to access the private event_id)
    def get_event_id(self):
        return self.__event_id

    # Method to create a new event
    def create_event(self):
        print(f"Event '{self.name}' created on {self.date} at {self.time} in {self.location}.")

    # Method to add an attendee
    def add_attendee(self, attendee_name):
        self.__attendees.append(attendee_name)
        print(f"{attendee_name} has been added to the event '{self.name}'.")

    # Method to remove an attendee
    def remove_attendee(self, attendee_name):
        if attendee_name in self.__attendees:
            self.__attendees.remove(attendee_name)
            print(f"{attendee_name} has been removed from the event '{self.name}'.")
        else:
            print(f"{attendee_name} was not found in the attendees list.")

    # Method to get the total number of attendees
    def get_total_attendees(self):
        return len(self.__attendees)

    # Method to display event details
    def display_event_details(self):
        print(f"Event: {self.name}, Date: {self.date}, Time: {self.time}, Location: {self.location}")
        print(f"Total Attendees: {self.get_total_attendees()}")


# Subclass for PrivateEvent
class PrivateEvent(Event):
    def __init__(self, event_id, name, date, time, location, invited_guests):
        super().__init__(event_id, name, date, time, location)
        self.invited_guests = invited_guests  # List of invited guests (only for private events)

    # Method to add an invited guest
    def add_invited_guest(self, guest_name):
        if guest_name not in self.invited_guests:
            self.invited_guests.append(guest_name)
            print(f"{guest_name} has been invited to the private event '{self.name}'.")
        else:
            print(f"{guest_name} is already on the invitation list.")

    # Method to display private event details
    def display_private_event_details(self):
        self.display_event_details()
        print(f"Invited Guests: {', '.join(self.invited_guests)}")


# Subclass for PublicEvent
class PublicEvent(Event):
    def __init__(self, event_id, name, date, time, location, open_for_public=True):
        super().__init__(event_id, name, date, time, location)
        self.open_for_public = open_for_public  # Indicates if the event is open to the public

    # Method to make the event open to the public
    def open_event_to_public(self):
        if not self.open_for_public:
            self.open_for_public = True
            print(f"The event '{self.name}' is now open to the public.")
        else:
            print(f"The event '{self.name}' is already open to the public.")

    # Method to display public event details
    def display_public_event_details(self):
        self.display_event_details()
        print(f"Open to Public: {'Yes' if self.open_for_public else 'No'}")


# Testing the Event Management System
if __name__ == "__main__":
    # Create instances of PrivateEvent and PublicEvent
    private_event = PrivateEvent(101, "Exclusive Gala", "2024-12-25", "18:00", "Hotel Royal", ["Alice", "Bob"])
    public_event = PublicEvent(102, "Community Fair", "2024-12-30", "10:00", "Central Park")

    # Create events
    private_event.create_event()
    public_event.create_event()

    # Add and remove attendees for private event
    private_event.add_attendee("Charlie")
    private_event.remove_attendee("Bob")
    private_event.add_attendee("David")

    # Add invited guests for private event
    private_event.add_invited_guest("Eve")
    private_event.add_invited_guest("Alice")  # Already invited
    private_event.display_private_event_details()

    # Add attendees for public event
    public_event.add_attendee("Frank")
    public_event.add_attendee("Grace")

    # Open the public event to the public
    public_event.open_event_to_public()

    # Display details of the public event
    public_event.display_public_event_details()

    # Display total attendees for each event
    print(f"Total Attendees for '{private_event.name}': {private_event.get_total_attendees()}")
    print(f"Total Attendees for '{public_event.name}': {public_event.get_total_attendees()}")
```

```python
# Base class for Flight
class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats):
        self.__flight_number = flight_number  # Private attribute: flight number is hidden using encapsulation
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = total_seats  # Private attribute: available seats

    # Getter for flight_number (to access the private flight_number)
    def get_flight_number(self):
        return self.__flight_number

    # Method to book a seat
    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked successfully on flight {self.get_flight_number()}.")
        else:
            print("No available seats on this flight.")

    # Method to cancel a reservation
    def cancel_reservation(self):
        if self.__available_seats < 50:  # Assuming the flight can have a max of 50 seats for simplicity
            self.__available_seats += 1
            print(f"Reservation cancelled successfully on flight {self.get_flight_number()}.")
        else:
            print("No reservation found to cancel.")

    # Method to get the remaining available seats
    def get_remaining_seats(self):
        return self.__available_seats

    # Method to display flight details
    def display_flight_details(self):
        print(f"Flight {self.get_flight_number()} from {self.departure_airport} to {self.arrival_airport}")
        print(f"Departure: {self.departure_time}, Arrival: {self.arrival_time}")
        print(f"Available Seats: {self.get_remaining_seats()}")

# Subclass for DomesticFlight
class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, baggage_allowance):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.baggage_allowance = baggage_allowance  # Specific attribute for Domestic flights

    # Method to display domestic flight details
    def display_domestic_flight_details(self):
        self.display_flight_details()
        print(f"Baggage Allowance: {self.baggage_allowance} kg")

# Subclass for InternationalFlight
class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, visa_required):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.visa_required = visa_required  # Specific attribute for International flights

    # Method to display international flight details
    def display_international_flight_details(self):
        self.display_flight_details()
        print(f"Visa Required: {'Yes' if self.visa_required else 'No'}")

# Testing the Airline Reservation System
if __name__ == "__main__":
    # Create instances of DomesticFlight and InternationalFlight
    domestic_flight = DomesticFlight("DL123", "Los Angeles", "New York", "2024-12-20 10:00", "2024-12-20 14:00", 100, 30)
    international_flight = InternationalFlight("AI456", "Mumbai", "London", "2024-12-25 23:30", "2024-12-26 05:00", 150, True)

    # Display flight details
    domestic_flight.display_domestic_flight_details()
    print()
    international_flight.display_international_flight_details()

    # Book seats on both flights
    domestic_flight.book_seat()
    international_flight.book_seat()

    # Cancel a reservation
    domestic_flight.cancel_reservation()

    # Get remaining seats
    print(f"Remaining seats on Domestic Flight: {domestic_flight.get_remaining_seats()}")
    print(f"Remaining seats on International Flight: {international_flight.get_remaining_seats()}")
```
